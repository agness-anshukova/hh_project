# Преобразование данных

Преобразование данных в столбце **"Образование и ВУЗ"**. 

В исследовании рассматривается уровень образования. Признак учебного заведения в данном случае не рассматривается как существенный.

```python
# Преобразование данных
education = ['высшее', 'неоконченное высшее', 'среднее специальное','среднее']
def fill_education(line):
    if line.find('Высшее') == 0:
        return education[0]
    elif line.find('Неоконченное') == 0:
        return education[1]
    elif line.find('Среднее специальное') == 0:
        return education[2]
    elif line.find('Среднее') == 0 and line.find('Среднее специальное') == -1:
        return education[3]
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(fill_education)
hh_df.drop(columns=['Образование и ВУЗ'], inplace=True)
```

Преобразование столбца **"Пол, возраст"**. В исходных данных он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Выделяем каждый параметр в отдельный столбец.

Создаем два новых признака **"Пол"** и **"Возраст"**. При этом:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

```python
# Преобразование данных
hh_df['Пол, возраст'] = hh_df['Пол, возраст'].apply(lambda line: line.split(' , ') if type(line) == str else line)
hh_df['Пол'] = hh_df['Пол, возраст'].apply(lambda lst: 'Ж' if lst[0] == 'Женщина' else 'М')
hh_df['Возраст'] = hh_df['Пол, возраст'].apply(lambda lst: int(lst[1].split(' ')[1]))
hh_df.drop(columns=['Пол, возраст'], inplace=True)
```

Преобразование признака **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца выделяем общий опыт работы соискателя в месяцах, новый признак будет иметь название "Опыт работы (месяц)".

При этом учитываем следующие особенности:
* В данном признаке есть пропуски. Если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* В данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* Не учитываем информацию, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* Опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

```python
# Преобразование данных

def crop_skill_line(line):
    if type(line) is np.nan:
        return np.nan
    elif type(line) is str:
        if line.strip() == 'Не указано':
            return np.nan
        if len(line) == 0:
            return np.nan

        if line.find('Опыт работы') != -1:
            line = line[line.find('Опыт работы')+len('Опыт работы'):]
            lst = line.split(' ')
            wrds = []
            for wrd in lst:
                if len(wrd) > 0 and wrd[0].isupper():
                    break
                else:
                    wrds.append(wrd)
    
            line = ' '.join(wrds).strip()

        
        # если не отработал код выше (должность или название месяца начинается с маленькой буквы)
        if len(line) > 20:
            lst = line.split(' ')
            wrds = []
            for index, wrd in enumerate(lst):       
                if index > 0 and lst[index-1].find('месяц') != -1:
                    break
                elif index > 0 and (lst[index-1].find('год') != -1 or lst[index-1].find('лет') != -1 ) and line.find('месяц') == -1 :
                    break
                else:
                    wrds.append(wrd)
    
            line = ' '.join(wrds).strip()

    return line

def process_skill(line):
    if type(line) is str:
        if line.find('месяц') != -1 and (line.find('год') == -1 and line.find('лет') == -1 ):
            line = line[:line.find('месяц')].strip()
            return int(line)
        elif line.find('месяц') == -1 and line.find('год') != -1:
            line = line[:line.find('год')].strip()
            return int(line)*12
        elif line.find('месяц') == -1 and line.find('лет') != -1:
            line = line[:line.find('лет')].strip()
            return int(line)*12
        elif line.find('месяц') != -1 and line.find('года') != -1:
            year = line[:line.find('года')].strip()
            month = line[line.find('года')+len('года'):line.find('месяц')].strip()
            return int(year)*12 + int(month) 
        elif line.find('месяц') != -1 and line.find('год') != -1:
            year = line[:line.find('год')].strip()
            month = line[line.find('год')+len('год'):line.find('месяц')].strip()
            return int(year)*12 + int(month) 
        elif line.find('месяц') != -1 and line.find('лет') != -1:
            year = line[:line.find('лет')].strip()
            month = line[line.find('лет')+len('лет'):line.find('месяц')].strip()
            return int(year)*12 + int(month) 
    return line
        
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(crop_skill_line)
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы (месяц)'].apply(process_skill)
hh_df.drop(columns=['Опыт работы'], inplace=True)
```

```python
# Ответы на контрольные вопросы
# Чему равен медианный опыт работы (в месяцах) в нашей таблице?
print(f'Медианный опыт работы (в месяцах) равен {hh_df['Опыт работы (месяц)'].median()} месяцев.')
```

Преобразование признака "Город, переезд, командировки". Информация в нем представлена в следующем виде: **<Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>**. В скобках указаны необязательные параметры строки. Например, можно встретить следующие варианты:

* Москва , не готов к переезду , готов к командировкам
* Москва , м. Беломорская , не готов к переезду, не готов к командировкам
* Воронеж , готов к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам

Создаем отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. 

При этом признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

Список городов-миллионников:
    
   <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    </code>
    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.

* Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False. Возможны несколько вариантов описания готовности к переезду в признаке "Город, переезд, командировки". Например:
    * … , готов к переезду , …
    * … , не готова к переезду , …
    * … , готова к переезду (Москва, Санкт-Петербург, Ростов-на-Дону)
    * … , хочу переехать (США) , …
    
Интересует только сам факт возможности или желания переезда.
* Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к командировкам в признаке "Город, переезд, командировки". Например:
    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …
    * … , не готов к командировкам , …
    
Интересует только сам факт готовности к командировке.

```python
# Преобразование данных
def process_city(origin_ln):
    lst = origin_ln.split(',')
    line = lst[0]
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    if line.find('(') != -1:
        line = line[:line.find('(')]
    if line.find('Москва') != -1 or line.find('Санкт-Петербург') != -1 :
        return line.strip()
    else:
        for city in million_cities:
            if line.find(city) != -1:
                return 'город-миллионник'
        return 'другие'

def proc_relocation(line):
    relocation_dict = {'готов': False, 'не готов': False}
    lst = line.split(',')
    for item in lst:
        # если в одном элементе встретили не готов и переезд
        if item.find('не ') != -1 and item.find('готов') != -1 and item.find('переезд') != -1:
            relocation_dict['не готов'] = True
        # если в одном элементе встретили готов и переезд
        elif item.find('не ') == -1 and item.find('готов') != -1 and item.find('переезд') != -1:
            relocation_dict['готов'] = True
        elif item.find('хочу переехать') != -1:
            relocation_dict['готов'] = True

    if relocation_dict['не готов'] is True:
        return False
    elif relocation_dict['готов'] is True:
        return True
    elif relocation_dict['не готов'] is False and relocation_dict['готов'] is False:
        return False
    return 'Другое'

def proc_business_trip(line):
    business_trip_dict = {'готов': False, 'не готов': False}
    lst = line.split(',')
    for item in lst:
        # если в одном элементе встретили не готов и командировк
        if item.find('не ') != -1 and item.find('готов') != -1 and item.find('командировк') != -1:
            business_trip_dict['не готов'] = True
        # если в одном элементе встретили готов и командировк
        elif item.find('не ') == -1 and item.find('готов') != -1 and item.find('командировк') != -1:
            business_trip_dict['готов'] = True

    if business_trip_dict['не готов'] is True:
        return False
    elif business_trip_dict['готов'] is True:
        return True
    elif business_trip_dict['не готов'] is False and business_trip_dict['готов'] is False:
        return False   
    return 'Другое'

hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(process_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(proc_relocation)
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(proc_business_trip)
hh_df = hh_df.drop(columns=['Город, переезд, командировки'])
```

```python
# Ответы на контрольные вопросы
# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
print(f'В Санкт-Петербурге проживают {round(hh_df['Город'].value_counts(normalize=True).loc['Санкт-Петербург']*100)} процентов соискателей.')
# Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
def process_row(row):
    return str(row['Готовность к переезду']) + str(row['Готовность к командировкам'])

hh_df['Готовность к перезду и командировкам'] = hh_df.apply(lambda row: process_row(row), axis=1)
print(f'Одновременно и к переезду, и к командировкам готовы {round(hh_df['Готовность к перезду и командировкам'].value_counts(normalize=True).loc['TrueTrue']*100)} процента соискателя.')
```

Преобразование признака заработной платы **"ЗП"**. 
Помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:
* 30000 руб.
* 50000 грн.
* 550 USD

Указываем заработную плату в единой валюте.
Данные о курсах валют возьмём с ресурса MDF.RU за период с 29.12.2017 по 05.12.2019, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов.

Создаём признак "ЗП (руб)" - заработная плата в рублях.

```python
# Преобразование данных

# Преобразование названия валюты в ISO
def process_curency(arg):
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    curr = arg.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency_dict[curr]
    
# Отделение суммы  
def process_sum(arg):
    salary = float(arg.split(' ')[0])
    return salary

# Пересчет сумм в валюту РФ
def process_salary(row):
    return row['ЗП (руб)']*(row['close']/row['proportion'])


#1. Обновить признак "Обновление резюме" из таблицы с резюме, 
# переведя его в формат datetime и выделив из него дату. 
#hh_df['Обновление резюме'] = hh_df['Обновление резюме'].apply(lambda line: line.split(' ')[0])
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date

# Считываем и обрабатываем таблицу с валютным курсом. Оставляем столбцы:
#- currency — наименование валюты в ISO-кодировке;
#- date — дата;
#- proportion — пропорция;
#- close — цена закрытия (последний зафиксированный курс валюты на указанный день).
currency_df = pd.read_csv('data/ExchangeRates.csv', sep=',')
currency_df = currency_df.drop(columns=['per','time','vol'])
currency_df['date'] = pd.to_datetime(currency_df['date'], dayfirst=True).dt.date

#2. Выделить из столбца «ЗП» сумму желаемой заработной платы и наименование валюты, 
# в которой она исчисляется. 
# Наименование валюты перевести в стандарт ISO согласно таблице выше.
hh_df['Валюта'] = hh_df['ЗП'].apply(process_curency)
hh_df['ЗП (руб)'] = hh_df['ЗП'].apply(process_sum)

# 3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты 
# Значение close для рубля заполнить единицей 1 (курс рубля самого к себе).
merged_df = pd.merge(
                      hh_df,
                      currency_df, 
                      left_on=['Обновление резюме','Валюта'], 
                      right_on=['date','currency'], 
                      how='left'
                    )
merged_df['close'] = merged_df['close'].fillna(1)
merged_df['proportion'] = merged_df['proportion'].fillna(1)

#4. Умножить сумму желаемой заработной платы на присоединённый курс валюты (close) и 
# разделить на пропорцию. 
# Результат занести в новый столбец «ЗП (руб)».
hh_df['ЗП (руб)'] = merged_df.apply(lambda row: process_salary(row),axis=1)
hh_df = hh_df.drop(columns=['ЗП','Валюта'])
```

1) Точки далеко внизу справа - это пожилые люди с минимальным опытом. Возможные причины: недавняя иммиграция, долгий перерыв в карьере или радикальная смена профессии в предпенсионном возрасте.
2) Точки высоко слева - это молодые люди с огромным опытом, стаж был указан неверно. Таких 7 человек.

6. В результате анализа обнаружены потенциальные выбросы в признаке **"Возраст"**. 
Построим распределение признака "Возраст" в **логарифмическом масштабе**. 
Находим выбросы с помощью метода z-отклонения и удалим их из данных. Используем логарифмический масштаб. 

```python
log_age = np.log(hh_df['Возраст'])
fig = plt.figure(figsize=(10, 7))
histplot = sns.histplot(log_age, bins=30)
histplot.axvline(log_age.mean(), color='k', lw=2)
print(log_age.skew())
```

Распределение имеет положительную (правостороннюю) асимметрию.

```python
def outliers_z_score(data, feature, sigma_l=3,sigma_r=3):
    x = np.log(data[feature])
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - sigma_l * sigma
    upper_bound = mu + sigma_r * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned


outliers, cleaned = outliers_z_score(hh_df, 'Возраст', sigma_r=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
display(outliers)
hh_df = hh_df.drop(outliers.index)
```
